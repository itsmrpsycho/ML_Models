### 3.1 : New Customer Dataset

```python
import pandas as pd
import numpy as np
temp = pd.read_csv("./bin/q3/new_customers.csv")
X = temp.to_numpy()
# 0 CustomerID, 1 Gender, 2 Age, 3 Annual Income (k$), 4 Spending Score (1-100)
X[np.where(X[:,1] == 'Male'), 1] = 0 # male = 0
X[np.where(X[:,1] == 'Female'), 1] = 1 # Female = 1
X = X[:,1:]
for i in range(len(X.T)):
    X[:,i] = np.divide(X[:,i]-np.min(X[:,i]),np.max(X[:,i])-np.min(X[:,i]))
```

```python
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

class hierarchialClusteringModel():
    def __init__(self):
        pass
    
    def linkages(self, X, linkage_method):
        return sch.linkage(X, method=linkage_method)

    def plotDendrogram(self, linkage_matrix):
        plt.figure(figsize=(6,3))
        dendrogram = sch.dendrogram(linkage_matrix)
        plt.xlabel('Data Points')
        plt.ylabel('Distance')
        plt.title('Hierarchical Clustering Dendrogram')
        return dendrogram

    def plotAllDendrograms(self,X):
        linkageMatrix1 = self.linkages(X,linkage_method="single")
        linkageMatrix2 = self.linkages(X,linkage_method="complete")
        linkageMatrix3 = self.linkages(X,linkage_method="average")
        linkageMatrix4 = self.linkages(X,linkage_method="centroid")
        linkageMatrix5 = self.linkages(X,linkage_method="ward")

        self.plotDendrogram(linkageMatrix1)
        plt.title("Linkage method = single")
        plt.gca().set_xticklabels([])
        plt.show()

        self.plotDendrogram(linkageMatrix2)
        plt.title("Linkage method = complete")
        plt.gca().set_xticklabels([])
        plt.show()

        self.plotDendrogram(linkageMatrix3)
        plt.title("Linkage method = average")
        plt.gca().set_xticklabels([])
        plt.show()

        self.plotDendrogram(linkageMatrix4)
        plt.title("Linkage method = centroid")
        plt.gca().set_xticklabels([])
        plt.show()

        self.plotDendrogram(linkageMatrix5)
        plt.title("Linkage method = ward")
        plt.gca().set_xticklabels([])
        plt.show()
```

```python
hc = hierarchialClusteringModel()
hc.plotAllDendrograms(X)
```

```python
# Varying features
hc = hierarchialClusteringModel()
hc.plotAllDendrograms(X[:,1:])
```
